```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def plot_signal(t, x):
    '''
    t - Time vector for a Signal
    x - Signal 
    '''
    plt.figure(figsize=(18, 6))
    plt.plot(t, x)
    plt.show()
```

# Amplitude Modulation
           
           
### Carrier Signal

A  - Amplitude

fc - Frequency

t  - Time vector of the signal

ct - Carrier signal


```python

A=2; 
fc=0.5; 
t=np.arange(0, 60, 0.1)
ct=A*np.sin(2*np.pi*fc*t); 
plot_signal(t, ct)
```

### Message Signal
                
m  - amplitude sensitivity

M  - Amplitude of message signal

fm - Frequency of message signal

phase - Phase of message signal

mt - Modulating message signal


```python

m = 1.0
M = A * m
fm=.05
phase=0
mt=M*np.cos((2*np.pi*fm*t)+phase)
plot_signal(t, mt)
```

### Amplitudde Modulation Computation

yt - Amplitude Modulated signal

```python
yt=(1+m*np.cos((2*np.pi*fm*t) + phase))*A*np.sin(2*np.pi*fc*t);
plot_signal(t, yt)
```

# Frequency Modulation

### Carrier Signal
Ac  - Amplitude of carrier signal

fc  - Frequency fo carrier signal

t   - Time vector for the signal

xc  - Carrier signal


```python
Ac=2; 
fc=40; 
t=np.arange(0, 1.0, 0.0001)
xc=Ac*np.cos(2*np.pi*fc*t); 
plot_signal(t, xc)
```

###  Message Signal
m  - modulation index

fm - Frequency of message signal

xm - Modulating message signal

```python
m = 1.
fm=4
xm=np.sin((2*np.pi*fm*t))
plot_signal(t, xm)
```

###  Computing Frequency modulation

yt - Frequency modulated signal

```python
yt = Ac*np.cos(2. * np.pi * (fc * t + m * np.sin(2*np.pi*fm*t)))
plot_signal(t, yt)
```

# Phase Modulation

### Carrier Signal

Ac - Amplitude of carrier signal

fc - Frequency fo carrier signal

t  - Time vector for the signal

phase - Phase of Carrier signal

ct - Carrier signal

```python
Ac=2; 
fc=2; 
t=np.arange(0, 3.0, 0.0001)
phase = 0
ct=Ac*np.sin(2*np.pi*fc*t + phase); 
plot_signal(t, ct)
```

### Message Signal

k  - amplitude sensitivity

fm - Frequency of message signal

Am - Amplitude of the message signal

mt - Modulating message signal

```python
k = 1.0
fm=3
Am = 2
mt=Am*np.sin(2 * np.pi*fm*t)
plot_signal(t, mt)
```

### Computing Phase Modulation

yt - Phase modulated signal

```python
yt=Ac*np.sin((2*np.pi*fc*t) + (k * mt + phase))
plot_signal(t, yt)
```
